### **What is gTTS**
gTTS (Google Text-to-Speech)is a Python library and CLI tool to interface with Google Translate text-to-speech API. ... The output of the converted text is stored in the form of speech in the tts variable. The tts. save function allows us to save the converted speech in a format that allows us to play sounds.

```python
# import packages
# !pip install gTTS
#!pip install PyPDF2
import gtts
from gtts import gTTS
import PyPDF2
```

#### _*Create Audiobook*_

```python
# function for creating audiobook
def audioBook(file, file_type, lang='en', tld='com.au'):
    
# converting .pdf file to audio
    if file_type.lower() == 'pdf':
        all_txt = [] # ---> here we will store all pdf text 
        
        pdfReader = PyPDF2.PdfFileReader(open(file, 'rb'))
        for page_num in range(pdfReader.numPages):
            text =  pdfReader.getPage(page_num).extractText()
            all_txt.append(str(text))
        
        clean_txt = ' '.join([x.replace('\n', '') for x in all_txt if '\n' in x]) # ---> removing '\n' from extracted text
        tts = gTTS(text=clean_txt, lang=lang, tld=tld)
        tts.save('Audiobuk.mp3')
        print("file converted successfully and downloaded as 'Audiobook.mp3', pls check you current working directory")        
        
            
    
    # converting text file to audio
    elif file_type.lower() == 'txt':
        with open(file) as f: 
            lines = f.readlines()
            
        tts = gTTS(text=lines, lang=lang, tld=tld)
        tts.save('Audiobuk.mp3')
        f.close()
        print("file converted successfully and downloaded as 'Audiobook.mp3', pls check you current working directory")
    
    else:
        print('file type not supported')
```

```python
audioBook('2 page Clean Code.pdf', 'pdf')
```
